```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from tqdm import tqdm
import pickle
```

```python
tqdm.pandas()
```

```python
train_data = pd.read_csv("/kaggle/input/commonlitreadabilityprize/train.csv")
test_data = pd.read_csv("/kaggle/input/commonlitreadabilityprize/test.csv")
```

```python
train_data.info()
```

```python
train_data["excerpt"].head()
```

```python
vectorization_model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
embeddings = vectorization_model.encode(train_data["excerpt"].array, normalize_embeddings=True)
```

```python
t_data, v_data, y_train, y_vaidate = train_test_split(embeddings, train_data["target"], random_state=42)
```

```python
model = LinearRegression()
model.fit(t_data, y_train)
```

```python
predictions = model.predict(v_data)
```

```python
mean_squared_error(y_vaidate, predictions, squared=False)
```

```python
binary_model = pickle.dumps(model)
```

```python
with open("model.pkl", "wb") as f:
    f.write(binary_model)
```
